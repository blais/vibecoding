```python
from IPython.display import display, HTML
from IPython.core.getipython import get_ipython

# Function to insert a cell at a specified index
def insert_cell_at_index(index, cell_content):
    ip = get_ipython()
    if ip is None:
        raise RuntimeError("Could not get IPython instance. Are you running in a Jupyter environment?")
    
    # Create a new cell object
    new_cell = ip.user_ns.get('In')[0].__class__(cell_content)
    
    # Insert the new cell at the specified index
    ip.user_ns['_'] = new_cell
    ip.set_next_input(cell_content)#, replace=True)

# Call the function to insert a cell at index 1
insert_cell_at_index(1, '2+2')

# Display sobme output to confirm the modification
print("Cell inserted at index 1")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display

# Generate some sample data
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

# Function to insert a Matplotlib plot as a cnew cell
def insert_plot_cell_at_index(index, fig):
    plt.close(fig)  # Close the figure to avoid duplicate display
    display(fig.canvas.figure)

# Create a Matplotlib figure and plot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Sine Wave')

# Call the function to insert the plot as a new cell at index 1
insert_plot_cell_at_index(3, fig)
```
